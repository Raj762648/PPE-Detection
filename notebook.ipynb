```python
# Load pretrained YOLOv8 nano model
model = YOLO("yolov8n.pt")

# Train model
model.train(
        data="/content/drive/MyDrive/Colab Notebooks/Data/archive/data.yaml",
        epochs=50,
        imgsz=640,
        batch=16,
        project="runs",
        name="ppe_yolov8"
        )
```

```python
# Model inference

model = YOLO("best.pt")

cap = cv2.VideoCapture(0)  # Webcam

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    annotated_frame = results[0].plot()

    cv2.imshow("PPE Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
